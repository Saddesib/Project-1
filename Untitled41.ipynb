# Dora the Explorer TV show Game 



The game is a simple "Hide and Seek" style interactive experience where the player is in a digital environment searching for hidden objects. The hidden objects being her three best friends Backpack, Monkey and MAps. The main components of the game are as follows:

1. **Objective**: The player's goal is to find hidden objects within a larger bush image displayed on the screen.

2. **Environment**: The game window consists of a defined play area with a visually appealing background represented by a bigger bush image, which conceals the hidden objects.

3. **Characters and Objects**:
   - **Player**: The player is represented by a small circle on the screen, which they can control using arrow keys.
   - **Dora**: A character displayed on the screen, adding an element of engagement.
   - **Hidden Objects**: These are strategically placed objects within the bush image that the player needs to locate.

4. **Gameplay**:
   - The player uses arrow keys to move their character and explore the bush area.
   - When the player is in close proximity to a hidden object and clicks on it, the object becomes visible, and the game notifies the player of their discovery.
   - The player aims to find all the hidden objects to win the game.

5. **Feedback and Progress**:
   - The game provides feedback by displaying a "Hurray! You found the object!" message each time the player discovers a hidden object.
   - The game also informs the player of the number of remaining hidden objects they need to find to win.

6. **Winning Condition**:
   - The player wins the game upon finding all the hidden objects within the bush.

7. **Visual Elements**:
   - The game features visual elements like images of the bush, Dora character, and hidden objects to enhance the player's experience and engagement.

In summary, the game offers a visually engaging "Hide and Seek" experience where the player interacts with the environment, uncovers hidden objects, and works towards the goal of finding all the objects to achieve victory. The addition of feedback and character elements adds to the fun and immersion of the gameplay.

```python
import pygame
import sys
import random

# Initialize Pygame
pygame.init()

# Constants
SCREEN_WIDTH = 1000  # Increased screen width
SCREEN_HEIGHT = 600
FPS = 60

# Initialize the game window
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("Hide and Seek Game")

clock = pygame.time.Clock()

# Load the bigger bush image and Dora image
bigger_bush_image = pygame.image.load(r"C:\Users\ss303\Downloads\df8vhl4-6d0e81b6-dfe3-45bb-acf9-084a47226ba6.png")
bigger_bush_image = pygame.transform.scale(bigger_bush_image, (400, 400))  # Resize the image

dora_image = pygame.image.load(r"C:\Users\ss303\Downloads\images.jpg")
dora_image = pygame.transform.scale(dora_image, (100, 100))  # Resize Dora image

# Load object images
object_images = [pygame.image.load(r"C:\Users\ss303\Downloads\download.png"), pygame.image.load(r"C:\Users\ss303\Downloads\download_2.jpg"), pygame.image.load(r"C:\Users\ss303\Downloads\download_3.png")]

# Define a function to place bushes and objects randomly
def place_objects():
    # Position the objects within the bigger bush
    bigger_bush_x = (SCREEN_WIDTH - 400) // 2
    bigger_bush_y = (SCREEN_HEIGHT - 400) // 2
    objects = []

    for _ in range(3):
        x = random.randint(bigger_bush_x + 20, bigger_bush_x + 380)
        y = random.randint(bigger_bush_y + 20, bigger_bush_y + 380)
        object_index = random.randint(0, len(object_images) - 1)
        objects.append((x, y, object_images[object_index]))  # Assign a random object image

    return (bigger_bush_x, bigger_bush_y), objects

# Place bigger bush and objects
bigger_bush_position, objects = place_objects()
object_visibility = [False] * len(objects)  # Initialize object visibility

# Player position and radius
player_x, player_y = SCREEN_WIDTH // 2, SCREEN_HEIGHT // 2
player_radius = 10

# Dora's position
dora_x = bigger_bush_position[0] - 120  # Adjusted position to be in front of the bush
dora_y = (SCREEN_HEIGHT - 100) // 2

# Counter for the number of found objects
found_objects = 0

# Game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.MOUSEBUTTONDOWN and event.button == 1:  # Left mouse button
            mouse_x, mouse_y = pygame.mouse.get_pos()
            for i, (obj_x, obj_y, obj_image) in enumerate(objects):
                distance = pygame.math.Vector2(mouse_x - obj_x, mouse_y - obj_y).length()
                if not object_visibility[i] and distance <= player_radius + 30:  # Object is invisible and player is near
                    object_visibility[i] = True
                    found_objects += 1
                    remaining_objects = len(objects) - found_objects
                    print(f"Hurray! You found the object! {remaining_objects} more to find to win.")

    # Handle player controls
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT]:
        player_x -= 5
    if keys[pygame.K_RIGHT]:
        player_x += 5
    if keys[pygame.K_UP]:
        player_y -= 5
    if keys[pygame.K_DOWN]:
        player_y += 5

    # Clear the screen
    screen.fill((255, 255, 255))

    # Draw the bigger bush
    screen.blit(bigger_bush_image, bigger_bush_position)

    # Draw objects
    for i, (obj_x, obj_y, obj_image) in enumerate(objects):
        if object_visibility[i]:
            screen.blit(obj_image, (obj_x - obj_image.get_width() // 2, obj_y - obj_image.get_height() // 2))

    # Draw Dora in front of the bush
    screen.blit(dora_image, (dora_x, dora_y))

    # Draw the player
    pygame.draw.circle(screen, (0, 0, 255), (player_x, player_y), player_radius)

    # Update the screen
    pygame.display.update()

    clock.tick(FPS)

pygame.quit()
sys.exit()
```
